```python
import pandas as pd
import numpy as np

df = pd.read_csv("all_matching_rules_exe.csv")
```

```python
df.head()
```

```python
import ast
h = set()
tot = 0
for v in df.yara_matching_rules.values:
    v = ast.literal_eval(v)
    tot += len(v)
    for x in v:
        h.add(x)
print(tot)
print(len(h))
```

```python
df['yara_matching_rules_2'] = df['yara_matching_rules'].apply(lambda x : ast.literal_eval(x))
```

```python
print("This takes several minutes")

for c in h:
    df[c] = np.zeros(df.shape[0])

print(df.shape)

for _, v in enumerate(df.yara_matching_rules_2):
    for c in h:
        if c in v:
            df.iloc[_, df.columns.get_loc(c)] = 1
print(df)
```

```python
del df['yara_matching_rules']
del df['yara_matching_rules_2']
```

```python
detection_dict = dict()
for c in h:
    detection_dict[c.split(":")[1][:49]] =  int(df[c].sum())
```

```python
from operator import itemgetter
lst = sorted(detection_dict.items(), key=itemgetter(1))
for t in lst: 
    print('%-50s %d' % (t[0], t[1]))
```
