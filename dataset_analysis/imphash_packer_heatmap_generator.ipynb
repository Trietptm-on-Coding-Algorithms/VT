```python
import pandas as pd
import os
""" For dll analysis."""
path = "./dll_analysis/packers"
file_name = "fixed_rules.csv"
file_to_scan = os.path.join(path, file_name)
""" For exe analysis. """
#path = "./exe_analysis/packers"
#file_name = "rules_fixed.csv"
#file_to_scan = os.path.join(path, file_name)
df1 = pd.read_csv(file_to_scan)
df1.head()
```

```python
df1 = df1[df1["yara_detected"]==True]
```

```python
df1.head()
```

```python
import ast
h = set()
tot = 0
for v in df1.yara_matching_rules.values:
    v = ast.literal_eval(v)
    tot += len(v)
    for x in v:
        h.add(x)
print(tot)
print(len(h))
```

```python
df1['yara_matching_rules_2'] = df1['yara_matching_rules'].apply(lambda x : ast.literal_eval(x))
```

```python
import numpy as np
print("This takes several minutes")

for c in h:
    df1[c] = np.zeros(df1.shape[0])

print(df1.shape)

for _, v in enumerate(df1.yara_matching_rules_2):
    for c in h:
        if c in v:
            df1.iloc[_, df1.columns.get_loc(c)] = 1
print(df1)
```

```python
del df1['yara_matching_rules']
del df1['yara_matching_rules_2']
```

```python
detection_dict = dict()
for c in h:
    detection_dict[c.split(":")[1][:49]] =  int(df1[c].sum())
```

```python
import re
from operator import itemgetter
lst = sorted(detection_dict.items(), key=itemgetter(1))
accepted_list = []
for t in lst: 
    if(t[1]>3 and re.match(".*(HasOverlay|HasDigital).*",t[0]) is None):
        accepted_list.append(t[0])
```

```python
print(accepted_list)
```

```python
df1 = pd.read_csv(file_to_scan)
df1.head()
df1['yara_matching_rules'] = df1['yara_matching_rules'].apply(lambda x : ast.literal_eval(x))
```

```python
df_dictionary = {el : pd.DataFrame() for el in accepted_list}
```

```python
import re
for index, row in df1.iterrows():
    for rule in accepted_list:
        assigned = False
        for el in row.yara_matching_rules:
            if(re.match(".*"+rule+".*",el)):
                assigned = True
                df_dictionary[rule]=df_dictionary[rule].append(row)
                break
        if(assigned == True):
            continue
            


    
```

```python
import gc
gc.collect()
```

```python
for rule in accepted_list:
    df_dictionary[rule]["Packer"]=rule
    df_dictionary[rule]=df_dictionary[rule].loc[:,["Exe name", "imphash", "Packer"]]
    
```

```python
for rule in accepted_list:
    print(df_dictionary[rule].shape)
gc.collect()
```

```python
for rule in accepted_list:
    mf = df_dictionary[rule]
    """ Performing cartesian product of the dataset with itself
    ( it will be useful to use the apply function over the same row of the dataframe).
    Each row will now contain two samples of the same family 
    (too complex and, probably, useless to perform it over all the samples, independently on the family)"""
    cp = mf.merge(mf, how='outer', on = "Packer")
    cp["IAT equality"] = cp.apply(lambda x: int(x["imphash_x"]==x["imphash_y"]), axis=1)
    df_dictionary[rule] = cp
gc.collect()
```

```python
print(path)
```

```python
import os
import matplotlib.pyplot as plt
for rule in accepted_list:
    if(df_dictionary[rule].shape[0]<2):
        continue
    df_dictionary[rule].head()
    df_dictionary[rule] = df_dictionary[rule].pivot(index= "Exe name_x", columns="Exe name_y", values="IAT equality")
    gc.collect()
    ax = plt.pcolor(df_dictionary[rule])
    ax.get_figure().savefig(path+"/heatmap/imphash/"+rule+"_heatmap", dpi=900)
    df_dictionary[rule].to_csv(path+"/heatmap/imphash/"+rule+"_heatmap.csv")
    plt.clf()
```
